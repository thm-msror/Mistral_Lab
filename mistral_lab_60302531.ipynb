### Load API key and helper function
Note: You can view or download the helper.py file by clicking on the "Jupyter" logo to access the file directory

```python
import os
from dotenv import load_dotenv

# Load variables from .env file
load_dotenv()

# Get the key
api_key = os.getenv("MISTRAL_API_KEY")
```

```python
import os
from mistralai import Mistral, UserMessage
def mistral(user_message, model="mistral-small-latest", is_json=False):
    model = "mistral-large-latest"
    client = Mistral(api_key=api_key)
    messages = [
        UserMessage(content=user_message),
        ]
    chat_response = client.chat.complete(model=model,
                                         messages=messages,)
    return chat_response.choices[0].message.content
```

```python
mistral("hello, what can you do?")
```

Information Extraction with JSON Mode

```python
medical_notes = """
A 60-year-old male patient, Mr. Johnson, presented with symptoms of 
increased thirst, frequent urination, fatigue, and unexplained weight loss. 
Upon evaluation, he was diagnosed with diabetes,confirmed by elevated blood sugar levels. 
Mr. Johnson's weight is 210 lbs. He has been prescribed Metformin to be taken twice daily with meals. 
It was noted during the consultation that the patient is a current smoker.
"""
```

```python
prompt = f"""
Extract information from the following medical notes:{medical_notes}
Return json format with the following JSON schema:
{{"age": {{"type": "integer"}},
"gender": {{"type": "string","enum": ["male", "female", "other"]}},
"diagnosis": {{"type": "string","enum": ["migraine", "diabetes", "arthritis", "acne"]}},
"weight": {{"type": "integer"}},
"smoking": {{"type": "string",
"enum": ["yes", "no"]}}}}
"""
```

```python
response = mistral(prompt, is_json=True)
print(response)
```
